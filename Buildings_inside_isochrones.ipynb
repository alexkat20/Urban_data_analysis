```python
import geopandas as gpd
import shapely
from shapely.geometry import Point
import folium
```

```python
walk = gpd.read_file('walk_isochrone.geojson')
houses = gpd.read_file("final_buildings.geojson")
houses = houses.to_crs(32634)
walk = walk.to_crs(houses.crs) # data - здания

# Check which polygons are inside the single polygon
polygons_inside_walk = houses[houses.geometry.within(walk.geometry.iloc[0])]
map = folium.Map(location = [54.722642, 20.514075], zoom_start = 13, tiles='CartoDB positron')

folium.GeoJson(data=walk, style_function=lambda x: {'fillColor': 'pink'}).add_to(map)
folium.GeoJson(data=polygons_inside_walk, style_function=lambda x: {'fillColor': 'green'}).add_to(map)
folium.Marker([54.722642, 20.514075]).add_to(map)
map
```

Изохрона 10 минутной пешеходной доступности

```python
iso = gpd.read_file('drive_isochrone.geojson')
iso = iso.to_crs(houses.crs) # data - здания

# Check which polygons are inside the single polygon
polygons_inside_drive = houses[houses.geometry.within(iso.geometry.iloc[0])]
map = folium.Map(location = [54.722642, 20.514075], zoom_start = 13, tiles='CartoDB positron')

folium.GeoJson(data=iso, style_function=lambda x: {'fillColor': 'pink'}).add_to(map)
folium.GeoJson(data=polygons_inside_drive, style_function=lambda x: {'fillColor': 'green'}).add_to(map)
folium.Marker([54.722642, 20.514075]).add_to(map)
map
```

```python
round(len(polygons_inside_drive)/len(polygons_inside_walk),2)
```

Изохрона 5-минутной доступности на личном автомобиле от выбранной точки.
Покрывает большую площадь, чем изохрона 10 минутной пешеходной доступности, в неё попало в 2 раза больше домов.

```python
iso = gpd.read_file('pt_isochrone.geojson')
routes = gpd.read_file('pt_routes.geojson')
stops = gpd.read_file('pt_stops.geojson')
iso = iso.to_crs(houses.crs) # data - здания
stops = stops.to_crs(houses.crs) # data - здания
routes = routes.to_crs(houses.crs) # data - здания

# Check which polygons are inside the single polygon
polygons_inside_pt = houses[houses.geometry.within(iso.geometry.iloc[0])]
map = folium.Map(location = [54.722642, 20.514075], zoom_start = 13, tiles='CartoDB positron')

folium.GeoJson(data=iso, style_function=lambda x: {'fillColor': 'pink'}).add_to(map)
folium.GeoJson(data=polygons_inside_pt, style_function=lambda x: {'fillColor': 'green'}).add_to(map)
folium.GeoJson(data=stops).add_to(map)
# folium.GeoJson(data=routes).add_to(map)
folium.Marker([54.722642, 20.514075]).add_to(map)
map
```
