```python
import sys
import folium
```

```python
import geopandas as gpd
```

```python
blocks = gpd.read_file("./data1/Blocks.geojson")

blocks["block_id"] = range(len(blocks))
blocks
```

```python
district = gpd.read_file("./data1/district.geojson")

district["district_id"] = range(len(district))
district
```

```python
blocks = blocks.to_crs(district.crs) 
```

```python
neighborhoods = blocks.sjoin(district, how="left")
neighborhoods
```

```python
neighborhoods = neighborhoods.drop(columns=["index_right","name","population","administrative_unit_id"])
```

```python
neighborhoods["district_id"] = neighborhoods["district_id"].fillna(neighborhoods["district_id"].median())
```

```python
neighborhoods = neighborhoods.rename(columns={"district_id":"administrative_unit_id"})
neighborhoods
```

```python
neighborhoods.to_file("final_blocks.geojson", driver="GeoJSON")
```

```python
from CityGeoTools.metrics.data import CityInformationModel as BaseModel

city_model = BaseModel.CityInformationModel(city_name="Kaliningrad", city_crs=4326)
print(city_model)
```

```python
city_model.update_layer("Blocks", "final_blocks.geojson")
```

```python
print(city_model.Blocks)
```

```python
city_model.update_layer("Buildings", "./data1/final_buildings.geojson")
```

```python
city_model.methods.spacematrix.message
```

```python
from CityGeoTools.metrics.calculations import spacematrix
blocks_morphotypes = spacematrix.Spacematrix(city_model).get_morphotypes()
blocks_morphotypes = gpd.GeoDataFrame.from_features(blocks_morphotypes).set_crs(4326)
```

```python
def style_fn(feature):
    colors = [
        'red', 'blue', 'green', 'orange', 'purple', 'pink', 'cadetblue', 'darkblue', 
        'lightgreen', 'darkred', 'yellow'
    ]
    return {'fillColor': colors[int(feature["properties"]["spacematrix_cluster"])]}

blocks_morphotypes_1 = blocks_morphotypes[blocks_morphotypes["administrative_unit_id"] == 1]
blocks_morphotypes_1 = blocks_morphotypes_1.dropna(subset=["spacematrix_cluster"])
map = folium.Map(location = [54.71, 20.51], zoom_start = 12, tiles='CartoDB positron')
geo_j = folium.GeoJson(
    data=blocks_morphotypes_1, 
    style_function=style_fn,
    tooltip=folium.features.GeoJsonTooltip(["FSI", "GSI", "L", "MXI", "OSR", "spacematrix_cluster"])).add_to(map)
geo_j.add_to(map)
map
```
